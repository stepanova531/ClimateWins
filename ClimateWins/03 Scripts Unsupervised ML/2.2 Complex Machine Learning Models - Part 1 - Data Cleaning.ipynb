## 2. Data wrangling

```python
# Remove all the columns containing the name of three cities: GDANSK, ROMA, TOURS which don't exist in Pleasant weather data set
# Create a pattern 
pattern=r'(GDANSK|ROMA|TOURS)'

# Filter the columns to remove
columns_to_remove=df.filter(regex=pattern).columns
```

```python
columns_to_remove.shape
```

```python
columns_to_remove
```

```python
# Remove the columns and create a new data set
df_drop=df.drop(columns=columns_to_remove, inplace=False)
```

```python
df_drop.shape
```

```python
# Check for missing values
df_drop.isnull().sum()
```

```python
# Remove all the columns containing the two observations: wind_speed and snow_depth  
# Create a pattern 
pattern_1=r'(wind_speed|snow_depth)'

# Filter the columns to remove
columns_to_remove_1=df_drop.filter(regex=pattern_1).columns 
```

```python
columns_to_remove_1.shape
```

```python
columns_to_remove_1
```

```python
# Remove the columns and create a new data set
df_drop_1=df_drop.drop(columns=columns_to_remove_1, inplace=False)
```

```python
df_drop_1.shape
```

```python
# Add three new columns cloud_cover, pressure and humidity for three locations and copy the values from the closest weather staions
df_drop_1[['KASSEL_cloud_cover', 'MUNCHENB_pressue', 'STOCKHOLM_humidity']]=df_drop_1[['LJUBLJANA_cloud_cover', 'SONNBLICK_pressure', 'OSLO_humidity']].copy()
```

```python
df_drop_1.shape
```

```python
df_drop_1.info()
```

```python
# Move the added columns within dataframe
cols=df_drop_1.columns.tolist()
```

```python
col_134=cols[134] # Column 134 = KASSEL_cloud_cover 
col_135=cols[135] # Column 135 = MUNCHENB_pressue 
col_136=cols[136] # Column 136 = STOCKHOLM_humidity

cols_reordered=(
    cols[:56] +         # Columns before the column with index 56 (up to and including index 55)
    [col_134] +         # Place the column 134 between columns 55 and 56
    cols[56:93] +       # Columns from index 56 up to (but not including) index 93
    [col_135]+          # Place the column 135 between columns 92 and 93
    cols[93:118]+       # Columns from index 93 up to (but not including) index 118
    [col_136]+          # Place the column 136 between columns 117 and 118
    cols[118:134]       # Columns after index 118 and up to 134 (but not included)
)
```

```python
df_drop_2=df_drop_1[cols_reordered]
```

```python
df_drop_2.shape
```

```python
df_drop_2.head()
```

```python
# Remove columns DATE and MONTH
df_drop_3=df_drop_2.drop(columns=['DATE', 'MONTH'], inplace=False)
```

```python
df_drop_3.shape
```

```python
df_drop_3.head()
```

```python
# Remove column DATE from Pleasnt weather data set
df_ans_drop=df_ans.drop(columns=['DATE'], inplace=False)
```

```python
df_ans_drop.shape
```

```python
df_ans_drop.head()
```

```python
# Export cleaned weather data set
df_drop_3.to_csv(os.path.join(path, 'Prepared Data', 'weather_cleaned.csv'))
```
